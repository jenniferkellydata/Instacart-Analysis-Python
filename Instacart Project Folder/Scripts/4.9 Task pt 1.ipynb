# step 5

Complete the fundamental data quality and consistency checks you’ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.

Quality and consistency check notes: 
- data type of user_id addressed above, check for mixed types in other columns and fix
- check for missing values and fix
- check for dupes and fix

```python
for col in df_cust.columns.tolist():
  weird = (df_cust[[col]].applymap(type) != df_cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cust[weird]) > 0:
    print (col)
```

```python
# changing first_name to object data type

df_cust['first_name'] = df_cust['first_name'].astype('object')
```

```python
# confirming

df_cust.info()
```

```python
# checking for missing values

df_cust.isnull().sum()
```

```python
# investigating missing first_name values

df_cust_nan = df_cust[df_cust['first_name'].isnull() == True]
```

```python
df_cust_nan
```

```python
# addressing missing values - replacing NaN with generic 'Customer' to eliminate NaNs

df_cust['first_name'].fillna('Customer', inplace=True)
```

```python
# checking if replacement successful

df_cust.isnull().sum()
```
