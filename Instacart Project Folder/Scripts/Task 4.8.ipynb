#4. Create a loyalty flag for existing customers using transform() and loc() functions.

```python
ords_prods_merge.groupby('user_id')
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

#6. Create a spending flag for each user based on the average price across all their orders.

```python
ords_prods_merge.groupby('user_id')
```

```python
ords_prods_merge['spend_habits'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['spend_habits']<10, 'spending_flag'] = 'Low Spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['spend_habits']>=10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

#7. Create a frequency flag that marks the regularity of a user's ordering behavior according to the median in the 'days_since_prior_order' column.

```python
ords_prods_merge.groupby('user_id')
```

```python
ords_prods_merge['ordering_behavior'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['ordering_behavior']>20, 'order_frequency_flag'] = 'Non-frequent Customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['ordering_behavior']>10) & (ords_prods_merge['ordering_behavior']<=20), 'order_frequency_flag'] = 'Regular Customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['ordering_behavior']<=10, 'order_frequency_flag'] = 'Frequent Customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
